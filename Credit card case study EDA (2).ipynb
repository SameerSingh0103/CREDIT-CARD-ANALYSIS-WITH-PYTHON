## 1. In the above dataset,

###  a. In case age is less than 18, replace it with mean of age values.

```python
Acq_spend_df.Age = np.where(Acq_spend_df.Age<18, round(Acq_spend_df.Age.mean()),Acq_spend_df.Age)
```

###  b. In case spend amount is more than the limit, replace it with 50% of that customerâ€™s limit. Customerâ€™s limit provided in acquisition table is the per transaction limit on his card

```python
Acq_spend_df.Amount= np.where(Acq_spend_df.Amount>Acq_spend_df.Limit,Acq_spend_df.Limit/0.50,Acq_spend_df.Amount)
```

###  c. Incase the repayment amount is more than the limit, replace the repayment with the limit

```python
Acq_Repayment_df.Amount=np.where(Acq_Repayment_df.Amount>Acq_Repayment_df.Limit,Acq_Repayment_df.Limit,Acq_Repayment_df.Amount)
```
